```python
import image_tools
import data_preparation
import keras
import matplotlib.pyplot as plt
import numpy as np
import skimage.io
```

```python
imgs, segs = data_preparation.load_data("JSRT_imgs", "JSRT_segs", relpath="./prepared_data/")
```

```python
image_tools.show_annotation(imgs[1], segs[1])
```

```python
model = keras.models.load_model('./gen2/seg_0.h5', compile=False)
```

```python
preds = model.predict(imgs)
```

```python
image_tools.show_annotation_pred(imgs[3], preds[3])
```
